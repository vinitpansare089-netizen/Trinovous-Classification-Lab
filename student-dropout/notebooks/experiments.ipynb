```python
# def convert_to_int(value):

# It should:

# return integer if valid
# return "Invalid input" if not

def convert_to_int(value):
    try:
        return int(value)
    except ValueError as e :
        return f'Invalid input => {e}' 

print(convert_to_int(10))
print(convert_to_int("vinit"))
```

```python
# 4. File loader
# def read_file(path):

# 👉 Handle:

# file not found
# empty file

def read_file(path):
    try:
        with open(path, 'r') as f:
            return f.read()
    except FileNotFoundError:
        return "file not found"
```

```python
# Upgrade:

# def safe_divide(a, b):

# 👉 Add:

# If result is integer → return int
# Else → return float

def safe_divide(a, b):
    try:
        a = int(a)
        b = int(b)
        result = a / b

        if result.is_integer():
            return int(result)
        else:
            return result
    except ZeroDivisionError:
        return "cannot divide by zero"
    except TypeError:
        return "Invalid Input"
    
print(safe_divide(10,2))
print(safe_divide(7,2))
```
