```python
# def convert_to_int(value):

# It should:

# return integer if valid
# return "Invalid input" if not

def convert_to_int(value):
    try:
        return int(value)
    except ValueError as e :
        return f'Invalid input => {e}' 

print(convert_to_int(10))
print(convert_to_int("vinit"))
```

```python
import pandas as pd

def load_data(path):
    try:
        print("data loading")
        df = pd.read_csv(path)
        return df
    except FileNotFoundError as e:
        raise FileNotFoundError(f'Invalid file :{e}')

```
